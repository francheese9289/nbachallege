# Nba Regular Season 2018-19 Data Challenge

Your task will be to take the dataset given, and create an analysis answering the following 10 questions. This project will again test your knowledge of pandas in order to find the answers needed given the data you are presented with.

```python
import pandas as pd
```

```python
nba_data = pd.read_csv("csv_files/nbastats2018-2019.csv", sep =',')
nba_data.head()
```

# What was the average age of player in the league?

```python
nba_player_avg_ages = nba_data['Age'].mean()
round(nba_player_avg_ages)
#Answer = 26
```

# What player scored the most points?

```python
sorted_nba_data = nba_data.sort_values('Points', kind = 'mergesort',ascending = False)

sorted_nba_data.reset_index(drop=True).head()
#Answer = James Harden
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
sorted_nba_blocks = nba_data.sort_values('Blocks', kind = 'mergesort',ascending = False)

sorted_nba_blocks.reset_index(drop=True).head()
#Very confused about the post player thing. there's no F/C column??
#Answer = Myles Turner 
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
#again, not sure how to tell what data is regular season????? 
#group by team, average win percentage
nba_ws_data = nba_data.groupby('Team', as_index = False).mean(numeric_only=True).round(decimals = 2)[['Team','WS']].sort_values('WS', ascending = False)
nba_ws_data.reset_index(drop=True).head()
#Answer = Milwaukee Bucks
```

# What player had the best 3-pt percentage? 

```python
nba_3pt_data = nba_data.sort_values('FG3%', ascending = False)[['Name','FG3%','FG3A']]
nba_3pt_data.reset_index(drop=True).head()
#Answer = Tie between Scott Machado, Jordan Sibert, Trevon Duval, Eric Moreland
```

# Who played the most minutes during the season

```python
nba_mp_data = nba_data.sort_values('MP', ascending = False)[['Name','Team','MP']]
nba_mp_data.reset_index(drop=True).head()
#Answer = Tie between Paul George and Bradley Beal
```

# What player given their player effiecency rating was the clutchest during the season?

```python
nba_per_data = nba_data.sort_values('PER', ascending = False)[['Name','Team','PER']]
nba_per_data.reset_index(drop=True).head()
#Answer = Zhou Qi
```

# What team had the youngest roster?

```python
nba_age_data = nba_data.groupby('Team', as_index = False).mean(numeric_only=True).round(decimals = 2)[['Team','Age']].sort_values('Age')
nba_age_data.reset_index(drop=True).head()
#Answer = Chicago Bulls
```

# Who is the highest paid player during the seasion?

```python
nba_data['Salary'] = nba_data['Salary'].replace('-', 0)
nba_data['Salary'] = nba_data['Salary'].astype(float)
nba_salary_data = nba_data.sort_values('Salary', ascending = False)[['Name','Team','Salary']]
nba_salary_data.reset_index(drop=True).head()
#Answer = Steph Curry
```

# At the end of a game, who WOULDN'T you want on the Free Throw Line?

```python
nba_ft_data = nba_data.sort_values(['FT%','FTA'],ascending = [True, False])[['Name','Team','FT%','FTA']]
nba_ft_data.reset_index(drop=True).head(20)
#Jacob Evans - has never attempted a freethow
#Could also be Tahjere McCall who has a 1.0 FTA but a 0% Ft%
```
